```python
#1)Unique Words in a Sentence
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Count of Unique words",len(unique_words))
print(unique_words)
```

```python
#2)Highest Salary from Employee Data
employees = {"Ravi": 75000,"Anita": 68000,"Kiran": 72000}
highest_salary = max(employees.values())
print("highest salary:",highest_salary)
```

```python
#3)Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
min_number = min(numbers)
max_number = max(numbers)
print("Minimum number:",min_number)
print("Maximum number:",max_number)
```

```python
#4)Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
    if price > 1000:
      count+=1
print("Products above 1000:",count)
```

```python
#5)Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
present_days = attendance.count("P")
total_days = len(attendance)
attendance_percentage = (present_days / total_days)*100
print("Attendance Percentage:",attendance_percentage)
```

```python
#6)Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9000124392, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:",unique_numbers)
```

```python
#7)Count Character Frequency
text="sathviksk"
char_count={}
for char in text:
    if char in char_count:
        char_count[char]+=1
    else:
        char_count[char]=1
print(char_count)
```

```python
#8)Convert List to Tuple
numbers= [1,3,3,5]
print(tuple(numbers))
```

```python
#9)Check if a Key Exists in a Dictionary
data = {"employee":"sathvik", "age":20}
if "employee" in data:
  print("employee exists")
else:
  print("employee does not exist")
```
