```python
#2) Analyze student results.
marks = [45, 78, 90, 33, 60]
for mark in marks:
    if mark>=50:
        print(mark,"Pass")
    else:
        print(mark,"Fail")
```

```python
#3) Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]
for name in names:
    clean_name = name.strip().lower()
    print(clean_name)

```

```python
#4) Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    if len(message) > 10:
        print(len(message),"Long message")
```

```python
#5) Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1
print(f"Total ERROR logs: {error_count}")
```
